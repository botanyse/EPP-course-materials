```python
import pandas as pd
from pathlib import Path
import plotly.express as px
import plotly.graph_objects as go

pd.options.mode.copy_on_write = True
pd.options.future.infer_string = True
pd.options.plotting.backend = "plotly"
```

```python
this_file_parent_dir = Path('.').resolve().parent
bld = this_file_parent_dir / "bld"
merged_data = pd.read_csv(bld / "merged_data.csv")
```

```python
score_names = {
    "antisocial": "bpiA",
    "anxiety": "bpiB",
    "headstrong": "bpiC",
    "hyperactive": "bpiD",
    "peer": "bpiE",
}
```

```python
merged_data.set_index(["childid","year"])
```

```python
def create_plot(df):
    fig  =  px.scatter(df, x="anxiety", y="bpiB", facet_col="age", trendline='ols')
    return go.Figure(fig)
```

```python
for key,value in score_names.items():
    print(key)
```

```python
create_plot(merged_data)
```
